```python
import os
import glob
import pandas as pd  # Import pandas for reading CSV files

# Get a list of all CSV files matching the pattern
file_list = glob.glob('experiment_*.csv')  # Renamed 'file' to 'file_list' to avoid confusion
print(f"✅ Found {len(file_list)} experiment CSV files:")
for file_path in file_list:  # Iterating through file paths
    print(file_path)  # Print the file path only once

df_list = []

# Loop through each CSV file
for i, file_path in enumerate(file_list, start=1):  # Using 'file_list'
    try:
        # Read the CSV into a pandas DataFrame
        df = pd.read_csv(file_path)
        # Add an experiment number column
        df["No"] = i
        # Append the DataFrame to the list
        df_list.append(df)
    except Exception as e:
        print(f"⚠️ Error reading {file_path}: {e}")

# Check if any DataFrames were loaded and print the shape of the first one if available

if df_list:
    merged_experiments = pd.concat(df_list, ignore_index=True)
    print("✅ Merging completed! Final dataset shape:", merged_experiments.shape)
else:
    print("❌ No valid experiment CSVs found to merge.")

merged_experiments.to_csv("merged_experiments.csv")
```

```python
import pickle
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Load the dataset
data = pd.read_csv('/content/final.csv')


# Label Encoding for categorical variables
label_encoders = {}
categorical_columns = ['tool_condition', 'machining_finalized', 'passed_visual_inspection']

for column in categorical_columns:
    le = LabelEncoder()
    data[column] = le.fit_transform(data[column])
    label_encoders[column] = le

    # Save each label encoder separately
    with open(f'label_encoder_{column}.pkl', 'wb') as file:
        pickle.dump(le, file)

print("Label encoders saved successfully!")
```
