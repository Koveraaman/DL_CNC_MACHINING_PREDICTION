```python
import os
import glob
import pandas as pd  # Import pandas for reading CSV files

# Get a list of all CSV files matching the pattern
file_list = glob.glob('experiment_*.csv')  # Renamed 'file' to 'file_list' to avoid confusion
print(f"✅ Found {len(file_list)} experiment CSV files:")
for file_path in file_list:  # Iterating through file paths
    print(file_path)  # Print the file path only once

df_list = []

# Loop through each CSV file
for i, file_path in enumerate(file_list, start=1):  # Using 'file_list'
    try:
        # Read the CSV into a pandas DataFrame
        df = pd.read_csv(file_path)
        # Add an experiment number column
        df["No"] = i
        # Append the DataFrame to the list
        df_list.append(df)
    except Exception as e:
        print(f"⚠️ Error reading {file_path}: {e}")

# Check if any DataFrames were loaded and print the shape of the first one if available

if df_list:
    merged_experiments = pd.concat(df_list, ignore_index=True)
    print("✅ Merging completed! Final dataset shape:", merged_experiments.shape)
else:
    print("❌ No valid experiment CSVs found to merge.")

merged_experiments.to_csv("merged_experiments.csv")
```

```python
from scipy import stats
def two_sample(d1,d2):
  t=0
  f=0
  for i in  range(31):
    sample1=d1.sample(frac=0.03)
    sample2=d2.sample(frac=0.03)
    t_test,p_value=stats.ttest_ind(sample1,sample2)
    if p_value < 0.055:
      f=f+1
    else:
      t=t+1
  if t>f:
    return True
  else:
    return False

#defining function for category vs category
def chisqare_cat_vs_cat(d1,d2):
  return True if stats.chi2_contingency(pd.crosstab(d1,d2))[1] < 0.055 else False

def annova_test(d1,d2):
  group=df[d2].unique()
  data={}
  for i in group:
    data[i]=df[d1][df[d2]==i]
  f_value,p_value=stats.f_oneway(*[i for i in data.values()])  #task for baby
  if p_value < 0.055:
    return False
  else:
    return True

final={}
for i in df.columns:
  final[i]={}
  for j in df.columns:
    if (i in continuous) and (j in continuous):
      result=two_sample(df[i],df[j])
    elif  (i in continuous) and (j in category):
      result=annova_test(i,j)
    elif (i in category) and (j in continuous):
      result=annova_test(j,i)
    elif (i in category) and (j in category):
      result=chisqare_cat_vs_cat(df[i],df[j])
    if result:
      final[i][j]=1
    else:
      final[i][j]=0
```

```python
# Define features and labels

# Split data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Normalize data
scaler = StandardScaler()
x_train_scaler = scaler.fit_transform(x_train)
x_test_scaler = scaler.transform(x_test)

# Build Deep Learning Model
model=Sequential()
model.add(Dense(1024,input_dim=x_train.shape[1],activation=tf.keras.layers.LeakyReLU(alpha=0.2)))
model.add(Dropout(0.2))
model.add(Dense(512,activation=tf.keras.layers.LeakyReLU(alpha=0.2)))
model.add(Dropout(0.23))
model.add(Dense(256,activation=tf.keras.layers.LeakyReLU(alpha=0.2)))
model.add(Dropout(0.15))
model.add(Dense(64,activation=tf.keras.layers.LeakyReLU(alpha=0.2)))
model.add(Dense(32,activation=tf.keras.layers.LeakyReLU(alpha=0.2)))
model.add(Dense(3,activation="softmax"))
model.compile(loss="categorical_crossentropy",optimizer="adam",metrics=["accuracy",tf.keras.metrics.F1Score(),tf.keras.metrics.Recall(),tf.keras.metrics.Precision()])


# Train Model
history = model.fit(x_train, y_train, epochs=20, batch_size=32, validation_data=(x_test, y_test))

# Evaluate Model
y_pred = model.predict(x_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_test_classes = np.argmax(y_test.values, axis=1)
print(classification_report(y_test_classes, y_pred_classes))
```
